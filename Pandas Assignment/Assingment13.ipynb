```python
import pandas as pd

data = { 
    "Name" : ["Anish","Sachin","Kartik","Soham","Omesh","Mohan","Rohit","Harish","karan","Rohan"],
    "RollNo" :[17,18,19,25,20,16,21,22,23,24],
    "Department" :["AIML","CS","IT","ECE","Civil","AIML","ENTC","IT","CS","CS"],
    "Salary":[56000,23000,44000,78000,35000,22000,42000,34000,46000,14000]
}
df = pd.DataFrame(data)
```

```python
df.sort_values(by="Salary")
```

```python
#descending order
df.sort_values(by="Salary", ascending=False)
```

```python
df["Department"].unique()
```

```python
#employess from each Dept
df["Department"].value_counts()
```

```python
df.groupby("Department")["Salary"].max()
```

```python
df.groupby("Department")["Salary"].min()
```

```python
df.sort_values(by="Department")
```
