```python
import pandas as pd
import numpy as np
data = { 
    "Name" : ["Anish","Sachin","Kartik","Soham","Omesh","Mohan","Rohit","Harish","karan","Rohan"],
    "Age" :[17,18,19,16,15,19,np.nan,22,23,24],
    "City" :["Nashik","Pune","Mumbai","Dubai","Londan",np.nan,"Paris","Brazil","France","Mumbai"],
    "Marks":[90,20,45,79,89,23,45,34,np.nan,89]
}
df = pd.DataFrame(data)

```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.fillna("NULL")
```

```python
df["Age"].fillna(df["Age"].mean())
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
#permenent removal of duplicates 
df =df.drop_duplicates()
```
