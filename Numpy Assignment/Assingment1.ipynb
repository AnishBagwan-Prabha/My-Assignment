```python
import numpy as np
marks = np.array([78, 85, 90, 67, 88, 92, 75, 81, 69, 95])
print("Marks Datatype:",marks.dtype)
print("Marks Array:",marks)
print("size:",marks.size)
```

```python
emps = np.array([101, 102, 103, 104, 105, 106])
print("Employee ids:",emps)
print("Dimensions:",emps.ndim)
print("Shape:",emps.shape)
```

```python
students = np.array([
    [78, 82, 90],
    [65, 70, 88],
    [91, 86, 79]
])
print("Array:",students)
print("Shape:",students.shape)
print("Size:",students.size)
print("Dtype:",students.dtype)
```

```python
zeros=np.zeros((2,2))
print("Zeors:",zeros)
ones=np.ones((5,5))
print("Ones :",ones)
identity = np.eye(6)
print("Identity:",identity)
```
